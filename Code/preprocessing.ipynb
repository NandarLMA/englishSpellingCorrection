```python
import cv2
def binarization(image):
    img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(img, (5, 5), 0)
    img = cv2.threshold(blur, 10,225,0)[1]
    #img = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    cv2.imwrite("Binarization.png", img)
    return img
```

```python
import numpy as np
def find_score(image,angle):
    data = inter.rotate(image, angle, reshape=False, order=0)
    hist = np.sum(data, axis=1)
    mean = np.sum(hist)/len(hist)
    score = np.sum((hist-mean) ** 2)
    return score
```

```python
import cv2
def removeNoise(image):
    img = np.asarray(image)
    converted_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    img = cv2.fastNlMeansDenoisingColored(converted_img, None, 10,10,7,15)
    cv2.imwrite("final.png",img)
    return img
```
